```python
import pandas as pd
import numpy as np
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score, 
                             confusion_matrix, roc_auc_score, classification_report, log_loss)

# Download NLTK stopwords and WordNetLemmatizer
nltk.download('stopwords')
nltk.download('wordnet')

# Initialize lemmatizer and stopwords
lemmatizer = WordNetLemmatizer()
stop_words = set(stopwords.words('english'))

# Text preprocessing function
def preprocess_text(text):
    # Lowercase the text
    text = text.lower()
    
    # Remove special characters, numbers, and punctuations
    text = re.sub(r'[^a-z\s]', '', text)
    
    # Tokenize and remove stopwords, and apply lemmatization
    words = text.split()
    clean_words = [lemmatizer.lemmatize(word) for word in words if word not in stop_words]
    
    return ' '.join(clean_words)

# Load the dataset
file_name = "/Users/manavpathak/Documents/Idmb sentiment data/Train.csv"
data = pd.read_csv(file_name)

# Apply preprocessing
data['cleaned_text'] = data['text'].apply(preprocess_text)

# Split the data
X = data['cleaned_text']
y = data['label']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)

# TF-IDF Vectorization
tfidf_vectorizer = TfidfVectorizer(max_features=5000)
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)

# Train Naive Bayes Classifier
naive_bayes = MultinomialNB()
naive_bayes.fit(X_train_tfidf, y_train)

# Predictions
y_pred = naive_bayes.predict(X_test_tfidf)
y_pred_prob = naive_bayes.predict_proba(X_test_tfidf)

# Metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')
#conf_matrix = confusion_matrix(y_test, y_pred)
roc_auc = roc_auc_score(y_test, y_pred_prob[:, 1])
log_loss_value = log_loss(y_test, y_pred_prob)

# Print metrics
print(f'Accuracy: {accuracy}')
print(f'Precision: {precision}')
print(f'Recall: {recall}')
print(f'F1 Score: {f1}')
#print(f'Confusion Matrix:\n{conf_matrix}')
print(f'AUC-ROC Score: {roc_auc}')
print(f'Log Loss: {log_loss_value}')
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix, roc_auc_score, log_loss, matthews_corrcoef, classification_report, roc_curve, auc
from sklearn.pipeline import Pipeline
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
import matplotlib.pyplot as plt

# Load the dataset
file_name = "/Users/manavpathak/Documents/Idmb sentiment data/Train.csv"
data = pd.read_csv(file_name)

# Better text preprocessing
nltk.download('stopwords')
nltk.download('wordnet')
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def preprocess_text(text):
    # Lowercasing and removing special characters
    text = text.lower()
    text = re.sub(r'\W', ' ', text)
    # Remove stopwords and apply lemmatization
    tokens = text.split()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stop_words]
    return ' '.join(tokens)

# Apply preprocessing
data['text'] = data['text'].apply(preprocess_text)

# Split the data
x = data['text']
y = data['label']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)

# TF-IDF Vectorization with optimization (max_features increased)
tfidf_vectorizer = TfidfVectorizer(max_features=10000)
x_train_tfidf = tfidf_vectorizer.fit_transform(x_train)
x_test_tfidf = tfidf_vectorizer.transform(x_test)

# KNN Classifier with Hyperparameter Tuning using GridSearchCV
pipeline = Pipeline([
    ('tfidf', TfidfVectorizer(max_features=10000)),
    ('knn', KNeighborsClassifier())
])

# Parameter grid for hyperparameter tuning
param_grid = {
    'knn__n_neighbors': [3, 5, 7, 9],
    'knn__weights': ['uniform', 'distance'],
    'knn__metric': ['euclidean', 'manhattan', 'cosine']
}

# GridSearchCV to find the best parameters
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', n_jobs=-1)
grid_search.fit(x_train, y_train)

# Best model after tuning
best_model = grid_search.best_estimator_
y_pred_knn = best_model.predict(x_test)
y_prob_knn = best_model.predict_proba(x_test)  # For Log Loss and ROC AUC

# Metrics
accuracy_knn_optimized = accuracy_score(y_test, y_pred_knn)
precision_knn_optimized = precision_score(y_test, y_pred_knn, average='weighted')
recall_knn_optimized = recall_score(y_test, y_pred_knn, average='weighted')
f1_knn_optimized = f1_score(y_test, y_pred_knn, average='weighted')
conf_matrix_knn_optimized = confusion_matrix(y_test, y_pred_knn)

# Additional Metrics
log_loss_knn = log_loss(y_test, y_prob_knn)
mcc_knn = matthews_corrcoef(y_test, y_pred_knn)
roc_auc_knn = roc_auc_score(y_test, y_prob_knn, multi_class='ovr')

# Print basic metrics
print(f'Best Parameters: {grid_search.best_params_}')
print(f'Accuracy (Optimized KNN): {accuracy_knn_optimized}')
print(f'Precision (Optimized KNN): {precision_knn_optimized}')
print(f'Recall (Optimized KNN): {recall_knn_optimized}')
print(f'F1 Score (Optimized KNN): {f1_knn_optimized}')
print(f'Confusion Matrix (Optimized KNN):\n{conf_matrix_knn_optimized}')

# Print additional metrics
print(f'Log Loss (Optimized KNN): {log_loss_knn}')
print(f'Matthews Correlation Coefficient (Optimized KNN): {mcc_knn}')
print(f'ROC AUC Score (Optimized KNN): {roc_auc_knn}')

# Classification Report
print(f'Classification Report:\n{classification_report(y_test, y_pred_knn)}')

# Plotting ROC curve
fpr = {}
tpr = {}
roc_auc = {}

for i in range(len(np.unique(y_test))):
    fpr[i], tpr[i], _ = roc_curve(y_test, y_prob_knn[:, i], pos_label=i)
    roc_auc[i] = auc(fpr[i], tpr[i])

# Plot each class's ROC curve
plt.figure(figsize=(8, 6))
for i in range(len(fpr)):
    plt.plot(fpr[i], tpr[i], label=f'Class {i} ROC curve (area = {roc_auc[i]:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Each Class')
plt.legend(loc='lower right')
plt.show()
```
